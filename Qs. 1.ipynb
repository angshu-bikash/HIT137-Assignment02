```python
import os
import pandas as pd

#specifying the folder/file path
file_path = "C:/Users/ASUS/Desktop/Semester 1, 2/HIT137 Software Now\Assignment 2/data & image/"

#initiating the list that will hold all the data
merge_text = []

#look for the .csv files and collect the necessary data
for file in os.listdir(file_path):
    if file.endswith('.csv'):
        df = pd.read_csv(os.path.join(file_path, file))

        # Check for 'TEXT' (csv 2,3,4) or 'SHORT-TEXT' (csv 1) column
        if 'TEXT' in df.columns:
            info = df['TEXT'].dropna().tolist()
        elif 'SHORT-TEXT' in df.columns:
            info = df['SHORT-TEXT'].dropna().tolist()
        
        # Add the text to the list, using extend to make all info individual in the merge_text
        merge_text.extend(info)

#combining in a single .txt file with the capability to handle different punctuations & other symbols (utf-8)
with open('combined_texts.txt', 'w', encoding='utf-8') as f:
    for text in merge_text:
        f.write(text + '\n')  

```

### Task 3.1: Finding & storing the top 30 most used used in the text collected from the .csv file 

```python
from collections import Counter
import re

# read from combined text file that we created in the 1st snippet
with open('combined_texts.txt', 'r', encoding='utf-8') as f:
    text_data = f.read()

#preprocess text: remove non-alphanumeric characters and split
# r is used so that any it doesn't interpret backslash (/) as escape character
# \b\w+\b matches with words surrounded by the boundaries
words = re.findall(r'\b\w+\b', text_data.lower())

#counting word occurrences
word_counts = Counter(words)

#collecting 30 most common words
top30_words = word_counts.most_common(30)

#top 30 words to a CSV file (two columns: Word, Count)
top_words_df = pd.DataFrame(top30_words, columns=['Word', 'Count'])
top_words_df.to_csv('top30_words.csv', index=False)
```

```python
#printing the top 30 for our reference
for word, count in top30_words:
    print(f"Word: {word}, Count: {count}")
```

### Task 3.2: Finding the word count using specific bio-task,tokenizer model

```python
from transformers import AutoTokenizer

#downloading the distributed 
bio_tokenizer = AutoTokenizer.from_pretrained("dmis-lab/biobert-base-cased-v1.1")
```

```python
from collections import Counter

#using the batch_size as have a bit lowe resource in local envioronment
#using max_lenngth as in some cases, the token number was exceeding 512 limit
batch_size = 1000 
max_length = 512   

#batching the tokenizers with truncation
def batch_tokenize(text_data, batch_size, max_length):
    tokens = []
    for i in range(0, len(text_data), batch_size):
        batch = text_data[i:i+batch_size]
        tokenized_batch = bio_tokenizer.tokenize(batch)
        #truncating the tokens when they areexceed length
        if len(tokenized_batch) > max_length:
            tokenized_batch = tokenized_batch[:max_length]
        tokens.extend(tokenized_batch)
    return tokens

#calling the function to start tokenizing
tokens = batch_tokenize(text_data, batch_size, max_length)

#counting the tokens of top 30
token_counts = Counter(tokens)
top_30_tokens = token_counts.most_common(30)

print(top_30_tokens)
```

### Finding NER using SciSpacy(en_ner_bc5cdr_md model) & BioBERT

#### Using SciSpacy

```python
import spacy

#using bc5cdr dataset which is a corpus of bio-entities
nlp = spacy.load("en_ner_bc5cdr_md")

#Processtext
doc = nlp(text_data)

# listing the found diseases and drugs entities
#here, drugs doesn't have separate label, it's included in the 'CHEMICAL' label
diseases = [ent.text for ent in doc.ents if ent.label_ == 'DISEASE']
drugs = [ent.text for ent in doc.ents if ent.label_ == 'CHEMICAL']

print(f"Total Diseases: {len(set(diseases))}")
print(f"Total Drugs: {len(set(drugs))}")
```

#### Using BioBERT

```python
from transformers import pipeline

#using the BioBERT NER pipeline
ner_pipeline = pipeline("ner", model="dmis-lab/biobert-base-cased-v1.1", tokenizer="dmis-lab/biobert-base-cased-v1.1")

#processtext
bio_ner_results = ner_pipeline(text_data)

#Extracting the diseases and drugs entities, but now we have separate label for drugs ('B_DRUG')
bio_diseases = [res['word'] for res in bio_ner_results if res['entity'] == 'B-DISEASE']
bio_drugs = [res['word'] for res in bio_ner_results if res['entity'] == 'B-DRUG']

print(f"BioBERT Total Diseases: {len(set(bio_diseases))}")
print(f"BioBERT Total Drugs: {len(set(bio_drugs))}")
```

### Comparing the result

```python
#finding the common entities (after removing duplications on each side) of each model
common_diseases = set(diseases).intersection(set(bio_diseases))
common_drugs = set(drugs).intersection(set(bio_drugs))

#the matching disease & drugs in both model
print(f"Common Diseases Detected: {len(common_diseases)}")
print(f"Common Drugs Detected: {len(common_drugs)}")

# Most common diseases and drugs
print(f"Most common diseases: {Counter(diseases).most_common(5)}")
print(f"Most common drugs: {Counter(drugs).most_common(5)}")
```
